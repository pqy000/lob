```python
import numpy as np
import matplotlib.pyplot as plt

# 加载三个npy文件
data_50 = np.load('50_2020.npy')
data_300 = np.load('300_2020.npy')
data_500 = np.load('500_2020.npy')

# 提取trend（最后一列）
trend_50 = data_50[:, -1]
trend_300 = data_300[:, -1]
trend_500 = data_500[:, -1]

# 统计不同trend标签的数量
trend_values_50, trend_counts_50 = np.unique(trend_50, return_counts=True)
trend_values_300, trend_counts_300 = np.unique(trend_300, return_counts=True)
trend_values_500, trend_counts_500 = np.unique(trend_500, return_counts=True)

# 找出所有唯一的trend标签
all_trend_values = np.union1d(np.union1d(trend_values_50, trend_values_300), trend_values_500)

# 创建空列表来保存每个数据集的trend计数，确保每个标签都出现
counts_50 = [trend_counts_50[np.where(trend_values_50 == value)[0][0]] if value in trend_values_50 else 0 for value in all_trend_values]
counts_300 = [trend_counts_300[np.where(trend_values_300 == value)[0][0]] if value in trend_values_300 else 0 for value in all_trend_values]
counts_500 = [trend_counts_500[np.where(trend_values_500 == value)[0][0]] if value in trend_values_500 else 0 for value in all_trend_values]
```

```python
# 2. 绘制trend饼图
bar_width = 0.25
x = np.arange(len(all_trend_values))

plt.figure(figsize=(12, 8))
plt.bar(x, counts_50, width=bar_width, label='Futures 50', edgecolor="black", color='lightblue')
plt.bar(x + bar_width, counts_300, width=bar_width, label='Futures 300', edgecolor="black", color='lightgreen')
plt.bar(x + 2 * bar_width, counts_500, width=bar_width, label='Futures 500', edgecolor="black", color='lightcoral')

plt.title('Trend Distribution for 50, 300, and 500 Data')
plt.xlabel('Trend')
plt.ylabel('Count')
plt.xticks(x + bar_width, all_trend_values)
plt.legend()
plt.grid(True)
plt.show()

```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# 50数据集的饼图
axs[0].pie(counts_50, labels=all_trend_values, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'gold', 'pink'])
axs[0].set_title('50 Data Trend Distribution')

# 300数据集的饼图
axs[1].pie(counts_300, labels=all_trend_values, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'gold', 'pink'])
axs[1].set_title('300 Data Trend Distribution')

# 500数据集的饼图
axs[2].pie(counts_500, labels=all_trend_values, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'gold', 'pink'])
axs[2].set_title('500 Data Trend Distribution')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 加载三个npy文件
data_50 = np.load('50_2020.npy')
data_300 = np.load('300_2020.npy')
data_500 = np.load('500_2020.npy')

# 提取trend（最后一列）
trend_50 = data_50[:, -1]
trend_300 = data_300[:, -1]
trend_500 = data_500[:, -1]

# 统计不同trend标签的数量
trend_values_50, trend_counts_50 = np.unique(trend_50,  return_counts=True)
trend_values_300, trend_counts_300 = np.unique(trend_300, return_counts=True)
trend_values_500, trend_counts_500 = np.unique(trend_500, return_counts=True)

# 找出所有唯一的trend标签
all_trend_values = np.union1d(np.union1d(trend_values_50, trend_values_300), trend_values_500)

# 创建空列表来保存每个数据集的trend计数，确保每个标签都出现
counts_50 = [trend_counts_50[np.where(trend_values_50 == value)[0][0]] if value in trend_values_50 else 0 for value in all_trend_values]
counts_300 = [trend_counts_300[np.where(trend_values_300 == value)[0][0]] if value in trend_values_300 else 0 for value in all_trend_values]
counts_500 = [trend_counts_500[np.where(trend_values_500 == value)[0][0]] if value in trend_values_500 else 0 for value in all_trend_values]

# 美化柱状图
bar_width = 0.25
x = np.arange(len(all_trend_values))

plt.figure(figsize=(12, 8))
plt.bar(x, counts_50, width=bar_width, label='50 Data', edgecolor="black", color='lightblue', alpha=0.7)
plt.bar(x + bar_width, counts_300, width=bar_width, label='300 Data', edgecolor="black", color='lightgreen', alpha=0.7)
plt.bar(x + 2 * bar_width, counts_500, width=bar_width, label='500 Data', edgecolor="black", color='lightcoral', alpha=0.7)

# 调整字体大小，增加标题、标签、图例的可读性
plt.title('Trend Distribution for 50, 300, and 500 Data', fontsize=20)  # 标题字体大小
plt.xlabel('Trend', fontsize=16)  # 横轴标签字体大小
plt.ylabel('Count', fontsize=16)  # 纵轴标签字体大小
plt.xticks(x + bar_width, all_trend_values, fontsize=14)  # x轴刻度字体大小
plt.yticks(fontsize=14)  # y轴刻度字体大小
plt.legend(fontsize=14)  # 图例字体大小
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 分析结果说明：
# 1. 从柱状图中可以看出，趋势标签为2（上涨）和-2（下跌）在三个数据集中是最多的。
# 2. 趋势标签1（轻微上涨）和-1（轻微下跌）的数量依次较多。
# 3. 趋势标签0（稳态）的数量最少，这表明期货数据在稳态的情况下较少。
# 4. 这反映出期货趋势预测任务是一个挑战性的任务，波动较大且趋势难以稳定。

# 美化饼图
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

colors = ['lightblue', 'lightgreen', 'lightcoral', 'gold', 'pink']

# 50数据集的饼图
axs[0].pie(counts_50, labels=all_trend_values, autopct='%1.1f%%', colors=colors, shadow=True, startangle=90, textprops={'fontsize': 14})
axs[0].set_title('50 Data Trend Distribution', fontsize=16)

# 300数据集的饼图
axs[1].pie(counts_300, labels=all_trend_values, autopct='%1.1f%%', colors=colors, shadow=True, startangle=90, textprops={'fontsize': 14})
axs[1].set_title('300 Data Trend Distribution', fontsize=16)

# 500数据集的饼图
axs[2].pie(counts_500, labels=all_trend_values, autopct='%1.1f%%', colors=colors, shadow=True, startangle=90, textprops={'fontsize': 14})
axs[2].set_title('500 Data Trend Distribution', fontsize=16)

plt.show()
```
